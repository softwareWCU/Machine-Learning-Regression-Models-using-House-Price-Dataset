```python
import pandas as pd

df = pd.read_csv('housing-price.csv')  # change name if needed
df.head()
```

```python
df.info()
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Convert YES / NO columns to numbers
yes_no_cols = [
    'mainroad', 'guestroom', 'basement',
    'hotwaterheating', 'airconditioning', 'prefarea'
]

df[yes_no_cols] = df[yes_no_cols].replace({'yes': 1, 'no': 0})
df.head()
```

```python
#Encode furnishingstatus
df = pd.get_dummies(df, columns=['furnishingstatus'], drop_first=True)
df.head()
```

```python
#Separate Features (X) and Target (y)
X = df.drop('price', axis=1)
y = df['price']
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train/Test separate
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
#Linear Regression (House Price Prediction)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

#R² Score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
r2

#RMSE (Error in Price Units)
import numpy as np
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
print(f"RMSE: {rmse}")
print(f"R Square Score: {r2}")

coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lr.coef_
})

coefficients.sort_values(by='Coefficient', ascending=False)
```

```python
#KNN REGRESSION
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

# R² Score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

# RMSE
import numpy as np
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"RMSE: {rmse}")
print(f"R Square Score: {r2}")
```

```python
#POLYNOMIAL REGRESSION
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_scaled)

from sklearn.model_selection import train_test_split
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_poly, y, test_size=0.2, random_state=42
)

poly_lr = LinearRegression()
poly_lr.fit(X_train_p, y_train_p)

y_pred = poly_lr.predict(X_test_p)

# R² Score
from sklearn.metrics import r2_score
r2 = r2_score(y_test_p, y_pred)

# RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test_p, y_pred))

print(f"RMSE: {rmse}")
print(f"R Square Score: {r2}")
```

```python
#DECISION TREE REGRESSION
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

# R² Score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

# RMSE
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"RMSE: {rmse}")
print(f"R Square Score: {r2}")
```
